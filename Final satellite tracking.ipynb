## Retreiving the positions of the satl mentioned above

```python
class getSatlPosn:
    obj = satlAboveHead()
    satlPosnData = []
    gps_position_data_from_above_satellite = []
    # This initialization would be done only if input is not None 
    # print("{} {}".format(type(obj.above_satellite_request.json()), obj.above_satellite_request.json()))
    # print(type(obj.above_satellite_request))

    if( bool(obj.above_satellite_request)): 
        gps_position_data_from_above_satellite = obj.above_satellite_request.json()['above']
    else:
        satlPosnData = None
        print("Value with None exists")
    
    # print(gps_position_data.keys())
    print(gps_position_data_from_above_satellite)
    print("There are {} Military Satellites on top of your head. ".format(obj.above_no_of_satellites)) # THis returns the count of military satellites
    
    url_str = "http://api.n2yo.com/rest/v1/satellite/positions/"
    latitude = str(33.496552592013906)
    longitude = str(76.52195669878829)
    leading_url = "0/30/&apiKey=SW2CQL-P78YQG-8MWP3V-50M7"

    print("\n Satellites are revolving on top of india: ")
    for i in gps_position_data_from_above_satellite:
        print("\n {}".format(i['satid']) )

        sat_id = str(i['satid'])
        final_url = url_str+sat_id+"/"+latitude+"/"+longitude+"/"+leading_url
        print(final_url+"\n")
        # Making the REST call
        req = requests.get(final_url)
        pprint.pprint(req.json())
        # print("\n")
        satlPosnData.insert(i['satid'],req.json())
        
    print(satlPosnData)
```
